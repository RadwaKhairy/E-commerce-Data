```python
import pandas as pd

df = pd.read_csv("E-commerce_Data2.csv")
df
```

```python
from sklearn.preprocessing import StandardScaler

df_scaled = pd.DataFrame()

df_scaled[df.columns] = StandardScaler().fit_transform(df[df.columns])

df_scaled
```

```python
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize': [30, 20]}, font_scale=1.2)
sns.heatmap(df.select_dtypes(include = np.number).corr() , cmap = 'crest' , annot = True)
```

```python
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize': [30, 20]}, font_scale=1.2)
sns.heatmap(df_scaled.select_dtypes(include = np.number).corr() , cmap = 'crest' , annot = True)
```
